```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plan = pd.read_excel('Modelo_maio.xlsx', index_col='Ano')
```

```python
fig, axs = plt.subplots(figsize=(10, 3))
axs.plot(plan['aguabranca'], c='mediumseagreen', label='Observado', linewidth=3)
axs.plot(plan['Modelo'], c='salmon', label='Modelo', linewidth=3, linestyle='--')
axs.set_xlim(1979, 2022)
axs.set_ylim(0, 700)
axs.set_xlabel('Anos no ms Maio')
axs.set_ylabel('Precipitao (mm)')
plt.legend()
axs.grid(True, linestyle='--')

plt.savefig('Serie_temporal.png', dpi=300, bbox_inches= 'tight')
```
